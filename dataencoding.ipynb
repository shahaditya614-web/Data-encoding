```python
# ==========================================
# CREATE EMPLOYEE DATASET
# ==========================================

data = {
    "Employee_ID":[501,502,503,504,505,506,507,508],
    "Employee_Name":["Rohan","Priya","Amit","Neha","Kiran","Meera","Vikas","Pooja"],
    "Department":["HR","IT","Finance","IT","Marketing","HR","Finance","Marketing"],
    "Gender":["Male","Female","Male","Female","Male","Female","Male","Female"],
    "City":["Ahmedabad","Surat","Rajkot","Vadodara","Ahmedabad","Surat","Rajkot","Vadodara"]
}

df = pd.DataFrame(data)

print("Original Employee Dataset")

display(df)
```

```python
# ==========================================
# LABEL ENCODING
# ==========================================

le = LabelEncoder()

label_df = df.copy()

label_df["Department"] = le.fit_transform(label_df["Department"])

print("Label Encoded Dataset")

display(label_df)
```

```python
# ==========================================
# LABEL ENCODER CLASSES
# ==========================================

print("Department Classes")

print(le.classes_)
```

```python
# ==========================================
# ONE HOT ENCODING
# ==========================================

one_hot = pd.get_dummies(
    df,
    columns=["Gender"],
    dtype=int
)

print("One Hot Encoded Dataset")

display(one_hot)
```

```python
# ==========================================
# BINARY ENCODING
# ==========================================

binary = ce.BinaryEncoder(cols=["City"])

binary_df = binary.fit_transform(df)

print("Binary Encoded Dataset")

display(binary_df)
```

```python
# ==========================================
# SHAPE COMPARISON
# ==========================================

print("Original Shape :", df.shape)

print("Label Encoding Shape :", label_df.shape)

print("One-Hot Encoding Shape :", one_hot.shape)

print("Binary Encoding Shape :", binary_df.shape)
```

```python
# ==========================================
# CONCLUSION
# ==========================================

print("="*60)
print("DATA ENCODING COMPLETED SUCCESSFULLY")
print("="*60)

print("""
1. Label Encoding converted Department into numeric values.
2. One-Hot Encoding converted Gender into separate binary columns.
3. Binary Encoding converted City into binary representation.
4. Different encoding techniques are used based on the machine learning algorithm and dataset.
""")
```
